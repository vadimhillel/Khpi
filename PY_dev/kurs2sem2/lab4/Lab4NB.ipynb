```python
# Task 2

""" 
Вводиться число n. Розставити 1 та 0 у шаховому порядку, починаючи 
з 1 в матриці розміром nхn, використовуючи слайсінг.
"""

def task_2(n: int) -> list[list]:
    arr = [[0] * n for _ in range(n)] # placeholders
    
    for i in range(n):
        for j in range(n):
            arr[i][j] = (i + j) % 2
    
    return arr

def task_2_np(n: int) -> np.ndarray:
    arr = np.zeros((n, n), dtype=int) # placeholders
    
    # Ones in even and odd rows and cols
    arr[::2, ::2] = 1
    arr[1::2, 1::2] = 1
    
    return arr

arr = task_2(5)
arr_np = task_2_np(5)

print("\n".join(" ".join(map(str, row)) for row in arr))
print(arr_np)
```

```python
# Task 3

""" 
Вводяться 4 числа n, m, r, c. Вивести масив розміру n на m, в якому 
в кожному r -му рядку і в кожному c-му стовпчику стоять 0, а інші елементи дорівнюють 1.
"""

def task_3(n, m, r, c) -> list[list]:
    arr = [[1] * m for _ in range(n)]  # placeholders
    
    # 0 in cols by c
    for i in range(n):
        for j in range(0, m, c):
            arr[i][j] = 0
    
    # 0 in rows by r
    for j in range(m):
        for i in range(0, n, r):
            arr[i][j] = 0
    
    return arr

def task_3_np(n, m, r, c) -> np.ndarray:
    arr = np.ones((n, m), dtype=int)  # placeholders
    
    # 0 in cols by c
    arr[::r, :] = 0
    
    # 0 in rows by r
    arr[:, ::c] = 0
    
    return arr

n = 7
m = 7
r = 2
c = 3

start = time.process_time()
arr = task_3(n, m, r, c)
end = time.process_time()
print("Via loops, stms etc: {:.10}".format(end - start))

start = time.process_time()
arr_np = task_3_np(n, m, r, c)
end = time.process_time()
print("Via numpy lib: {:.10f}".format(end - start))

print("\n".join(" ".join(map(str, row)) for row in arr))
print(arr_np)
```

```python
# Task 17

""" 
Заповнити вектор значеннями від 0 до n. Замінити 
знаки для всіх значеннь, що менші за n/2 та більші за 3n/4 на протилежні
"""

def task_17(n) -> list[list]:
    vector = [_ for _ in range(n + 1)]  # placeholder

    for i in range(n+1):
        if n//4 < vector[i] < 3*n//4:
            vector[i] = 0

    return vector

def task_17_np(n) -> np.ndarray:
    vector = np.arange(n+1)  # placeholder

    mask = (vector > n//4) & (vector < 3*n//4)
    vector[mask] = 0

    return vector

vec = task_17(10)
vec_np = task_17_np(10)
print(vec)
print(vec_np)
```

3. Знайдемо $x_1, x_2, x_3$ $x_4$ використовуючи правило Крамера:

$$
x_1 = \frac{{\text{det}(A_1)}}{{\text{det}(A)}} = \frac{-7}{-7} = 1
$$

$$
x_2 = \frac{{\text{det}(A_2)}}{{\text{det}(A)}} = \frac{-7}{-7} = 1
$$

$$
x_3 = \frac{{\text{det}(A_3)}}{{\text{det}(A)}} = \frac{-7}{-7} = 1
$$

$$
x_4 = \frac{{\text{det}(A_4)}}{{\text{det}(A)}} = \frac{-7}{-7} = 1
$$

```python
class Testing:
    def __init__(self, A, b):
        self.A = A
        self.b = b

    # Method 1: Matrix Multiplication @
    def test_1(self):
            det_A = np.linalg.det(self.A)
            res = np.empty(len(self.A)) 
            
            for i in range(len(self.A)):
                A_ = self.A.copy()
                A_[:, i] = self.b
                res[i] = np.linalg.det(A_) / det_A
            return res

    # Method 2: Inverse Matrix A-1
    def test_2(self):
            A_inv = np.linalg.inv(self.A)
            return np.dot(A_inv, self.b)

    # Method 3: numpy.linalg.solve()
    def test_3(self):
            return np.linalg.solve(self.A, self.b)

def main():
    A = np.array([[1, 1, -1, -1],
                [0, 1, 2, -1],
                [1, -1, 0, -1],
                [-1, 3, -2, 0]])

    b = np.array([0, 2, -1, 0])

    solver = Testing(A, b)
    
    check_functions = {
    "Checking with matrix multiplication": solver.test_1,
    "Checking with inverse matrix": solver.test_2,
    "Checking with numpy.linalg.solve()": solver.test_3
    }
    
    if all(np.allclose(test(), check_functions["Checking with numpy.linalg.solve()"]()) for test in check_functions.values()):
        print("\n".join("{}: {}".format(k, check_functions[k]()) for k in check_functions))

if __name__ == "__main__":
    main()
```

### 3. Обчислити значення матричного виразу
- **а)** з використанням універсальних функцій бібліотеки NumPy.
- **b)** за допомогою ітеративних конструкцій (з використанням циклів, спискових включень тощо).
- **с)** для обох випадків підрахувати час виконання скрипту та зробити висновок.

#### Порівняти всі рішення за допомогою функції numpy.allclose()

```python
class Arr_exp_solver:
    """A(A^2 - B) - 2(B + A)B."""
    
    def __init__(self, A, B) -> None:
        self.A = A
        self.B = B
        
    def expression_via_np(self) -> np.ndarray:
        A_2 = np.dot(self.A, self.A) # A @ A
        res = np.dot(self.A, A_2 - self.B) - 2 * (self.B + self.A) @ self.B
        
        return res
        
    def operation(self, A, operation, B):
        """+, -, *, scalar * of matrices provided"""
        arr = [[0 for _ in range(3)] for _ in range(3)]
        
        if operation == '+': # -> add
            for i in range(3):
                for j in range(3):
                    arr[i][j] = A[i][j] + B[i][j]
        elif operation == '-': # -> subtract
            for i in range(3):
                for j in range(3):
                    arr[i][j] = A[i][j] - B[i][j]
        elif operation == '*': # -> multiply
            for i in range(3):
                for j in range(3):
                    for k in range(3):
                        arr[i][j] += A[i][k] * B[k][j]
        elif operation == 'r*': # -> scalar multiply
            scalar = A
            for i in range(3):
                for j in range(3):
                    arr[i][j] = scalar * B[i][j]
        return arr
    
    def expression(self):
        A_squared = self.operation(self.A, '*', self.A) # A^2
        
        # Full expression ->
        part1 = self.operation(self.A, '*', self.operation(A_squared, '-', self.B))
        part2 = self.operation(2, 'r*', self.operation(self.operation(self.B, '+', self.A), '*', self.B))
        res = self.operation(part1, '-', part2) # <- res of A(A^2 - B) - 2(B + A)B
        
        return res
    
def main():
    A = np.array([[2, 3, 1],
                [-1, 2, 4],
                [5, 3, 0]])

    B = np.array([[2, 7, 13],
                [-1, 0, 5],
                [5, 13, 21]])

    my_solver = Arr_exp_solver(A, B)
    
    start = time.time()
    exp_np = my_solver.expression_via_np()
    end = time.time()
    print("Via numpy lib: {:.6f}".format(end - start))
    
    start = time.time()
    exp = my_solver.expression()
    end = time.time()
    print("Via loops, stms etc: {:.6f}".format(end - start))
    
    if np.allclose(exp, exp_np):
        print("\nVia numpy lib:")
        print(my_solver.expression_via_np())

        print("\nVia loops, stms etc:")
        print("\n".join(" ".join(map(str, row)) for row in my_solver.expression()))

if __name__ == "__main__":
    main()
    
```

```python
from IPython.display import display, Javascript

class ExecutionTimeCs:
    def __init__(self):
        self.start_time = None

    def __enter__(self):
        self.start_time = time.process_time()

    def __exit__(self, exc_type, exc_value, traceback):
        end_time = time.process_time()
        elapsed_time = end_time - self.start_time
        print(f"Execution time: {elapsed_time:.6f} seconds")

def exec_time(cell):
    with ExecutionTimeCs():
        display(Javascript(cell))
```
